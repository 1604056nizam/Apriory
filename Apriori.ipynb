```python
association_rules = apriori(records, min_support=0.0960, min_confidence=0.2, min_lift=3, min_length=2) 
association_results = list(association_rules)

print(association_results[0])

for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python
data = pd.read_csv('Copy of Depression and Happiness Factor Analysis.csv')
data_new = data.iloc[:, 1:].copy()  # copy all data excluding 1st columnt
# print(data_new)
# print(data_new.isnull())  # Checking for null value

# Preprocessing
info = data_new.to_numpy().tolist()  # Making a list of data
# print(info)

# Convert all values to string format
for values in info:
    for i in range(len(values)):
        s = ''
        if type(values[i]) != str:
            s = str(values[i])
            values[i] = s


tr = TransactionEncoder()
# Encoding data to boolean values of array
t_info = tr.fit(info).transform(info)
# print(t_info)

df = pd.DataFrame(t_info, columns=tr.columns_)  # create a dataframe
# print(df)

# Using Apriori algo to create frequent itemsets
frequent_items = apriori(df, min_support=0.60, use_colnames=True)
print(frequent_items)
```

```python
result = association_rules(
    frequent_items, metric='confidence', min_threshold=.60)

print(result)
```

```python
new_result = result.loc[:, ['antecedents',
                            'consequents', 'support', 'confidence']]
new_result
```
